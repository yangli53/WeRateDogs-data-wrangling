```python
consumer_key = 'HIDDEN'
consumer_secret = 'HIDDEN'
access_token = 'HIDDEN'
access_secret = 'HIDDEN'

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)

api = tweepy.API(auth)
```

```python
tweet_ids = list(twitter['tweet_id'])
```

```python
# Query Twitter API for each tweet's json data
count = 0
fails_dict = {}
start = timer()
# Save each tweet's JSON as a new line in txt file 
with open ('tweet_json.txt', 'w') as outfile:
    for tweet_id in tweet_ids:
        count += 1
        print(str(count) + ':' + str(tweet_id))
        try:
            tweet = api.get_status(tweet_id, tweet_mode='extended')
            print('Success')
            json.dump(tweet._json, outfile)
            outfile.write('\n')
        except tweepy.TweepError as e:
            print('Fail')
            fails_dict['tweet_id'] = e
            pass
end = timer()
print(end - start)
print(fails_dict)
```

```python
# Load json data
data = []
with open ('tweet_json.txt') as json_file:
    for line in json_file:
        data.append(json.loads(line))
```

```python
# List of dictionaries to build and convert to a DataFrame later
tweet_list = []
for i in range(len(data)):
    tweet_id = data[i]['id']
    retweet_count = data[i]['retweet_count']
    favorite_count = data[i]['favorite_count']
    
    tweet_list.append({'tweet_id': tweet_id,
                       'retweet_count': retweet_count,
                       'favorite_count': favorite_count})
```

```python
tweet = pd.DataFrame(tweet_list, columns=['tweet_id', 'retweet_count', 'favorite_count'])
tweet.sample(5)
```

```python
twitter_clean = twitter.copy()
image_clean = image.copy()
tweet_clean = tweet.copy()
```

##### `twitter` table

#### Define

 - Dog stage columns (*pupper*, *puppo*, *doggo*, *floofer*) can be combined to 1 column with 4 variables.
 - Dog stage columns *pupper*, *puppo*, *doggo*, *floofer* have missing data. They cannot be retrieved so I will replace them as null. 
 - Data type of dog stage column should be category.

#### Code

I was trying to use .melt() to combine the 4 columns to 1 and drop the 3 duplicates based on tweet_id. But I found that some rows have 2 dog stage values. Thus, I decide to create a new column *dog_stage* and concatenate the values from the 4 dog stage columns. Then I will replace the values with the accurate stage names.

```python
# Create a new column and concatenate the values from the 4 dog stage columns
twitter_clean['dog_stage'] = twitter_clean.doggo.str.cat(twitter_clean.floofer).str.cat(twitter_clean.pupper).str.cat(twitter_clean.puppo)
```

```python
twitter_clean.dog_stage.value_counts()
```

```python
# Replace the values with the accurate stage names
vals_to_replace = {'NoneNoneNoneNone': 'None', 
                   'NoneNonepupperNone': 'pupper', 
                   'doggoNoneNoneNone': 'doggo', 
                   'NoneNoneNonepuppo': 'puppo', 
                   'doggoNonepupperNone': 'doggo,pupper',
                   'NoneflooferNoneNone': 'floofer',
                   'doggoflooferNoneNone': 'doggo,floofer', 
                   'doggoNoneNonepuppo': 'doggo,puppo'}
twitter_clean.dog_stage = twitter_clean.dog_stage.map(vals_to_replace)
twitter_clean.dog_stage.value_counts()
```

```python
# Replace "None" as null
twitter_clean.dog_stage.replace('None', np.nan, inplace=True)

# Drop the 4 dog stage columns
twitter_clean = twitter_clean.drop(['doggo', 'floofer', 'pupper', 'puppo'], axis=1)

# Change data type to category
twitter_clean.dog_stage = twitter_clean.dog_stage.astype('category')
```

#### Code

I notice that the invalid names are usually lowercased so I will create a list to include all the invalid names and replace them with "Unknown". 

```python
from collections import Counter
```

```python
# Find all invalid dog names
name_list = []
for name in twitter_clean.name:
    if name.islower():
        name_list.append(name)
Counter(name_list)
```

```python
# Replace them with "Unknown"
for name in twitter_clean.name:
    if name in name_list:
        twitter_clean.name = twitter_clean.name.replace(name, 'Unknown')
```

```python
# Replace "None" with "Unknown"
twitter_clean.name = twitter_clean.name.replace('None', 'Unknown')
```

#### Define

 - Some predictions are not accurate. I will create a function to get the correct predictions from the 3 prediction columns (*p1*, *p2*, *p3*).

#### Code

Create a function to get the correct predictions and create new columns *dog_breed* and *p_conf*.

```python
# Create a function
breed_list = []
conf_list = []

def prediction(row):
    if row['p1_dog'] == True:
        breed_list.append(row['p1'])
        conf_list.append(row['p1_conf'])
    elif row['p2_dog'] == True:
        breed_list.append(row['p2'])
        conf_list.append(row['p2_conf'])
    elif row['p3_dog'] == True:
        breed_list.append(row['p3'])
        conf_list.append(row['p3_conf'])
    else:
        breed_list.append('Unidentified')
        conf_list.append(0)

# Apply function 
image_clean.apply(prediction, axis=1)

# Create new columns
image_clean['dog_breed'] = breed_list
image_clean['p_conf'] = conf_list
```

After reviewed the data, I'm interested in the following 3 questions:

#### Question 1: Are dogs with higher ratings also retweeted and favorited more? Is retweet count correlated with favorite count? 

Some ratings are very high compared to others. I will consider them as outliers and exclude those with a rating over 20 for analysis.

```python
# Exclude those with very high ratings
twitter = twitter.query('rating_numerator <= 20')
```

```python
# Plot a scatter chart to see the correlation of rating_numerator and retweet_count
twitter.plot(x='rating_numerator', y='retweet_count', kind='scatter', alpha=1/10)
plt.xticks(np.arange(21))
plt.xlabel('Rating')
plt.ylabel('Retweet Count')
plt.title('Rating vs. Retweet Count');
```

According to the chart above, dogs with higher ratings tend to have more retweets.

```python
# Plot a scatter chart to see the correlation of rating_numerator and favorite_count
twitter.plot(x='rating_numerator', y='favorite_count', kind='scatter', alpha=1/10)
plt.xticks(np.arange(21))
plt.xlabel('Rating')
plt.ylabel('Favorite Count')
plt.title('Ratings vs. Favorite Count');
```

Similar results are observed here: dogs with higher ratings tend to have more favorites.

```python
plt.figure(figsize = [10, 5])

plt.subplot(1, 2, 1)
plt.scatter(data=twitter, x='retweet_count', y='favorite_count', alpha=1/10)
plt.xlabel('Retweet Count')
plt.ylabel('Favorite Count')
plt.title('Retweet Count vs. Favorite Count');

plt.subplot(1, 2, 2)
plt.scatter(data=twitter, x='retweet_count', y='favorite_count', alpha=1/10)
plt.xlim(0,40000)
plt.xticks(rotation=15)
plt.xlabel('Retweet Count');
```

There is a strong positive correlation between retweets and favorites. I "zoom in" the data to easily observe the relationship.

#### Question 2: Do puppies get higher ratings and more retweets and favorites than adult dogs. 

```python
twitter.groupby('dog_stage')['rating_numerator'].mean().sort_values(ascending=False)
```

```python
dog_stages = ['puppo', 'floofer','doggo', 'pupper']
ratings = []
retweets = []
favorites = []
for stage in dog_stages:
    sub = twitter[twitter.dog_stage == stage]
    ratings.append(sub.rating_numerator.mean())
    retweets.append(sub.retweet_count.mean())
    favorites.append(sub.favorite_count.mean())
```

```python
# Create a bar chart to visualize the results
plt.bar(x=dog_stages, height=ratings, width=0.5, alpha=0.8)
plt.title('Average Rating by Dog Stage')
plt.xlabel('Dog Stage')
plt.ylabel('Rating');
```

The average ratings of puppos, floofers and doggos are very close at 12 while the average rating of pupper is 10~11.

```python
twitter.groupby('dog_stage')['retweet_count'].mean().sort_values(ascending=False)
```

```python
# Create a bar chart to visualize the results
plt.bar(x=dog_stages, height=retweets, width=0.5, alpha=0.8)
plt.title('Average Retweet Count by Dog Stage')
plt.xlabel('Dog Stage')
plt.ylabel('Retweet Count');
```

Doggos are retweeted the most with an average retweets of 7648.

```python
twitter.groupby('dog_stage')['favorite_count'].mean().sort_values(ascending=False)
```

```python
# Create a bar chart to visualize the results
plt.bar(x=dog_stages, height=favorites, width=0.5, alpha=0.8)
plt.title('Average Favorite Count by Dog Stage')
plt.xlabel('Dog Stage')
plt.ylabel('Favorite Count');
```

Puppos are favorited the most with an average favorites of 20530.
