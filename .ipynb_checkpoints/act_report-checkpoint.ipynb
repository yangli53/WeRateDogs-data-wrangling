# Data Wrangling Project - WeRateDogs 

The dataset I analyzed is the tweet archive of Twitter user @dog_rates, also known as WeRateDogs. WeRateDogs is a Twitter account that rates people's dogs with a humorous comment about the dog. These ratings almost always have a denominator of 10. The numerators are almost always greater than 10. 11/10, 12/10, 13/10, etc. Why? Because "they're good dogs Brent." WeRateDogs has over 4 million followers and has received international media coverage.

![WeRateDogs](img/image_1.png)

The data I have contains basic tweet data for 2038 tweets including each tweet's text, rating, dog name, dog stage (i.e. doggo, floofer, pupper, and puppo), predicted dog breed from dog image, prediction confidence, retweet count and favorite count. 

After I reviewed the data, I'm interested in the following 3 questions:

1. Dog ratings are provided by owners so they are more subjective. Twitter retweets and favorites are generated by others so they are more objective. I want to know if dogs with higher ratings are also retweeted and favorited more. I also want to know if retweet count is correlated with favorite count.
2. I want to know if puppies get higher ratings and more retweets and favorits than adult dogs. 
3. I want to know which dog breed gets the highest rating and which breed are retweeted and favorited the most. 

#### Question 1: Are dogs with higher ratings also retweeted and favorited more? Is retweet count correlated with favorite count? 

Some ratings are very high compared to others. I just considered those no more than 20 in this analysis. I found that **dogs with higher ratings tend to have more retweets**.

![1](img/chart_1.png)

Similar results were observed in favorite count: **dogs with higher ratings tend to have more favorites**.

![2](img/chart_2.png)

Then I plotted a scatter chart to see the relationship between retweet count and favorite count. **A strong positive correlation was observed between retweet count and favorite count**. The one on the right is a "zoom-in" chart to easily visualize the relationship.

![3](img/chart_3.png)

#### Question 2: Do puppies get higher ratings and more retweets and favorites than adult dogs. 

There are 4 dog stages in this dataset: puppo, floofer, doggo and pupper. I consider puppers as puppies. Though I thought people might provide higher ratings for puppies (they are cute!) but the data showed that **puppers only get an average rating of 10~11 while puppos, floofers and doggos get higher average ratings of 12**.

![4](img/chart_4.png)

**Puppers get the fewest retweets of 2296 while doggos are retweeted the most with an average retweets of 7648**.

![5](img/chart_5.png)

**Puppers also get the fewest favorites of 6635 while puppos are favorited the most with an average favorites of 20530**.

![6](img/chart_6.png)

#### Question 3: Which dog breed gets the highest rating and which breed is retweeted and favorited the most?

**Bouvier Des Flandres** gets the highest average rating of 13, followed by Saluki, Tibetan Mastiff, Briard and Border Terrier.

![7](img/chart_7.png)

![Bouview Des Flandres](img/image_2.png)

**Standard Poodle** is the most retweeted dog breed with an average retweets of 7994, followed by Bedlington Terrier, Afghan Hound, French Bulldog and Samoyed.

![8](img/chart_8.png)

![Standard Poodle](img/image_3.png)

**Bedlington Terrier** is the most favorited dog breed with an average favorites of 23795, followed by Saluki, French Bulldog, Bouvier Des Flandres and Black-And-Tan Coonhound.

![9](img/chart_9.png)

![Bedlington Terrier](img/image_4.png)
